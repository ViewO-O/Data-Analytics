# Missing Values

```python
flights = pd.read_csv("C:/Users/User/Desktop/DA/flights.csv")
flights.tail()
```

```python
# เลือกข้อมูลที่มี missing value อย่างน้อย 1 ตัวในแต่ละแถว
flights[flights.isnull().any(axis=1)].head()
```

```python
# เลือก arr_delay ที่มีค่า missing value
flights1 = flights[ flights['arr_delay'].isnull( )]
flights1.head()
```

```python
# เลือกข้อมูลทั้งหมดที่ตัด missing value ออกไป
flights2_nomissing = flights.dropna()
flights2_nomissing.head(10)
```

```python
# เปลี่ยนค่า missing values ของ dep_delay ด้วยค่าศูนย์
print("any missing", flights[ flights['dep_delay'].isnull( )].any())
print("any missing after fill missing",flights['dep_delay'].fillna(0).isnull().any())
#nomiss =flights['dep_delay'].fillna(0)
#nomiss.isnull().any() 
# ตรวจสอบว่า nomiss เป็น null หรือไม่ จะให้ค่า True or False ถ้าเป็น null จะมีค่าเป็น true
```

```python
# เปลี่ยนค่า missing values ของ dep_delay ด้วยค่าเฉลี่ยของ dep_delay
mean_dep_delay = flights['dep_delay'].mean() #หรือจะใช้คำสั่ง mean_dep_delay = np.mean(flights['dep_delay'])
average_nomiss =flights['dep_delay'].fillna(mean_dep_delay) #หรือจะใช้คำสั่ง flights['dep_delay'].fillna(flights['dep_delay'].mean())
```

```python
mean_dep_delay = flights['dep_delay'].mean()
print(mean_dep_delay)
```
