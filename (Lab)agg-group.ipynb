# Data slicing and grouping

```python
#Group data using rank  #แสดงผลข้อมูลที่ได้จากการแยกตาม rank
#df.groupby(['rank']).groups
print(df.groupby(['rank']).ngroups)
df.groupby(['rank']).mean(numeric_only=True)
```

```python
#แสดงผลข้อมูลที่ได้จากการแยกตาม rank และ sex
print(df.groupby(['rank', 'sex']).ngroups)
df.groupby(['rank', 'sex']).mean(numeric_only=True) #แสดงผลค่าเฉลี่ยที่ได้จากการแยกตาม rank และ se
```

```python
df.mean(numeric_only=True)
```

```python
#Calculate mean of all numeric columns for the grouped object
df.groupby(['rank'], sort=False).mean(numeric_only=True)
```

```python
df_rank = df.groupby(['rank'], sort=False)
df_rank.mean(numeric_only=True)
```

```python
df.groupby('sex').mean(numeric_only=True)
```

```python
#Calculate the mean salary for men and women.  
df.groupby('sex')['salary'].mean(numeric_only=True)
```

```python
# ถ้าใช้วงเล็บก้ามปูที่ attribute salary ผลลัพธ์จะแสดงออกมาเป็น dataframe
df.groupby('sex')[['salary']].mean(numeric_only=True)
```

```python
# Group using 2 variables - sex and rank:
df.groupby(['rank','sex'], sort=True)[['salary']].mean(numeric_only=True)
```
